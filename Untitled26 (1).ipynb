```python
%matplotlib inline
import matplotlib.pyplot as plt
import math as m
import numpy as np
from skimage import img_as_float, img_as_ubyte
from skimage.io import imread, imsave, imshow
from skimage.color import rgb2yuv, yuv2rgb, gray2rgb
from scipy import signal
import skimage
img = imread('https://stepik.org/media/attachments/lesson/61037/tiger-gray-small.png')
imshow(img)
```

```python
# для изучения частот изображения
def Furie (img):
    freq = np.log(1+abs(np.fft.fftshift(np.fft.fft2(img))))
    return freq
```

задание 1: Построение гауссовской пирамиды:

```python
def Gauss_piramid (img, sigma, n_layers):
    Gauss_img = []
    Gauss_img.append(img)
    for i in range(1, n_layers + 1):
        Gauss_img.append(Gauss_filter(Gauss_img[i-1], sigma))
    return Gauss_img
```

```python
# наложение ядра
def Gauss_filter(img, sigma):
    core = Gauss_core(sigma)
    res_img = np.zeros(shape = img.shape)
    if len(img.shape) == 2:
        res_img = signal.convolve2d(img, core, mode = "same")
    else:
        res_img[:,:,0] = signal.convolve2d(img[:,:,0], core, mode = "same")
        res_img[:,:,1] = signal.convolve2d(img[:,:,1], core, mode = "same")
        res_img[:,:,2] = signal.convolve2d(img[:,:,2], core, mode = "same")
    res_img = (res_img)
    return res_img
```

```python
# Построение гауссовского ядра
def Gauss_core(sigma):   
    K = round(sigma*6+1)
    core = [[0] * K for i in range(K)]
    S = 0
    for x in range(-(K//2),K//2+1):
        for y in range(-(K//2),K//2+1):
            S += Gauss(sigma,x,y)
    for x in range(-(K//2),K//2+1):
        for y in range(-(K//2),K//2+1):
            core[x+K//2][y+K//2] = Gauss(sigma,x,y)/S
    return core
```

```python
# подсчет гаусс-функции
def Gauss(sigma,x,y):
    result = 1/((2*m.pi)*sigma**2)*m.exp((-x**2 - y**2)/(2*sigma**2))
    return result
```

```python
def visual(piramid, sigma, n_layers):
    fig = plt.figure(figsize=(16, n_layers))
    axes = np.zeros((2, n_layers+1), dtype=np.object)
    for i in range(n_layers+1):
        axes[0, i] = fig.add_subplot(2, n_layers+1, 1+i)
        axes[0, i].imshow(piramid[i], cmap='gray')
        axes[0, i].set_title('layer = '+str(i))
        axes[0, i].axis('off')
        axes[1, i] = fig.add_subplot(2, n_layers+1, n_layers+2+i)
        axes[1, i].imshow(Furie(piramid[i]), cmap='gray')
    plt.show()
```

Задание 2, аналогичные дейстия для лаплассовой пирамиды. 

```python
def Laplas_piramid (img, sigma, n_layers):
    Gauss_img = Gauss_piramid(img, sigma, n_layers)
    Laplas_img = []
    for i in range(1, n_layers + 1):
        Laplas_img.append(Gauss_img[i-1] - Gauss_img[i])
    Laplas_img.append(Gauss_img[n_layers])
    return Laplas_img
```
