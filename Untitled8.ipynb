```python
%matplotlib inline
from skimage import img_as_float, img_as_ubyte
from skimage.io import imread, imshow, imsave
import numpy as np
```

```python
a = 'https://stepik.org/media/attachments/lesson/58182/02.png'
img = imread(a)
```

```python
imshow(img)
print(img.shape)
```

```python
img_b = img[0:img.shape[0]//3,0:img.shape[1]]
print(img_b.shape)
img_b = img_as_float(img_b)
x = int(img_b.shape[0]*0.1)
y = int(img_b.shape[1]*0.1)
img_b  = img[y:(img.shape[0]//3-y),x:(img.shape[1]-x)]
imshow(img_b)
print(img_b.shape)
print(x,y)
```

```python
img_g = img[img.shape[0]//3:2*img.shape[0]//3,0:img.shape[1]]
print(img_g.shape)
img_g = img_as_float(img_g)
x = int(img_g.shape[0]*0.1)
y = int(img_g.shape[1]*0.1)
img_g  = img[img.shape[0]//3+y:2*img.shape[0]//3-y,x:(img.shape[1]-x)]
imshow(img_g)
print(img_g.shape)
print(x,y)
```

```python
img_r = img[2*img.shape[0]//3:img.shape[0],0:img.shape[1]]
print(img_r.shape)
img_r = img_as_float(img_r)
x = int(img_r.shape[0]*0.1)
y = int(img_r.shape[1]*0.1)
img_r  = img[2*img.shape[0]//3+y:img.shape[0]-y, x:(img.shape[1]-x)]
imshow(img_r)
print(img_r.shape)
```

```python
ab = img_b.shape[0]
ag = img_g.shape[0]
ar = img_r.shape[0]
if (ab != ag):
    img_g  = img[img.shape[0]//3+y:2*img.shape[0]//3-1-y,x:(img.shape[1]-x)]
if (ab != ar): 
    img_r  = img[2*img.shape[0]//3+y:img.shape[0]-1-y, x:(img.shape[1]-x)]
print(ab,ag,ar)
```

```python
max_r = 0
x_max_r = -15
y_max_r = -15
for x in range(-15,15):
    for y in range(-15,15):
        r = np.roll(img_r,y, axis=0)
        r = np.roll(r,x, axis=1)
        correlation = (r*img_g).sum()
        if max_r <= correlation:
            max_r = correlation
            x_max_r = x
            y_max_r = y
print(x_max_r,y_max_r)
```

```python
max_b = 0
x_max_b = -15
y_max_b = -15
for x in range(-15,15):
    for y in range(-15,15):
        b = np.roll(img_b,y, axis=0)
        b = np.roll(b,x, axis=1)
        correlation = (b*img_g).sum()
        if max_b <= correlation:
            max_b = correlation
            x_max_b = x
            y_max_b = y
print(x_max_b,y_max_b)
```
