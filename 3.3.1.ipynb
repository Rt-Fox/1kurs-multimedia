```python
%matplotlib inline
from skimage.io import imread, imsave, imshow
from skimage import img_as_float,img_as_ubyte
from numpy import array_equal,histogram, clip, dstack
```

Разделение фотографии на три канала, подсчет среднего значение для каждого из каналов

```python
img = imread('https://stepik.org/media/attachments/lesson/60610/railroad.png')
imshow(img)
img_f = img_as_float(img)
r = img_f[:,:,0]
g = img_f[:,:,1]
b = img_f[:,:,2]
mid_r=0
mid_g=0
mid_b=0
size = img.shape
print(size)
for i in range(size[0]):
    for k in range(size[1]):
        mid_r += r[i][k]
        mid_g += g[i][k]
        mid_b += b[i][k]
mid_r=mid_r/r.size
mid_g=mid_g/r.size
mid_b=mid_b/r.size
print(mid_r,mid_g,mid_b)
```

Подсчет коэффициентов "серого мира"

```python
AVG = (mid_r+mid_g+mid_b)/3
print(AVG)
rw = mid_r/AVG
gw = mid_g/AVG
bw = mid_b/AVG
print(rw,gw,bw)
```

Изменение каналов изображения и их обьединение:

```python
r = (clip(r/rw,0,1))
g = (clip(g/gw,0,1))
b = (clip(b/bw,0,1))
img = dstack((r, g, b))
img = img_as_ubyte(img)
imshow(img)
```
