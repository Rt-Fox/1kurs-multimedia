```python
%matplotlib inline
import math as m
from skimage import img_as_float, img_as_ubyte
from skimage.io import imread, imsave, imshow
from scipy import signal
import numpy as np
import copy
img = imread('https://stepik.org/media/attachments/lesson/61037/tiger-gray-small.png')
imshow(img)
```

```python
arr = [[-1/10, -2/10, -1/10] for i in range(3)]
arr[1][0] = -2/10
arr[1][1] = 22/10
arr[1][2]=-2/10
out_img = np.zeros(shape = (img.shape[0] - 2, img.shape[1] - 2))
for x in range(img.shape[0] - 2):
    for y in range(img.shape[1] - 2):
        out_img[x, y] = np.clip(np.sum(img[x:x + 3, y:y + 3]*arr), 0, 255)
out_img=out_img.astype('uint8')
imshow(out_img.astype('uint8'))            
```

```python
omg = imread('https://stepik.org/media/attachments/lesson/61037/unsharp-tiger.png')
imshow(omg)
print(out_img-omg)
np.array_equal(out_img, omg)
```
