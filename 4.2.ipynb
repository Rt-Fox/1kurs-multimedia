```python
%matplotlib inline
import math as m
from skimage.io import imread, imsave, imshow
from scipy import signal
import numpy as np
import copy
img = imread('https://stepik.org/media/attachments/lesson/61041/tiger-gray-small.png')
imshow(img)
img.shape
```

```python
out_img = np.zeros(shape = (img.shape[0] - 6, img.shape[1] - 6))
for x in range(img.shape[0] - 6):
    for y in range(img.shape[1] - 6):
        out_img[x, y] =np.percentile(img[x:x + 7, y:y + 7], 50, interpolation='nearest')
out_img = out_img.astype('uint8')
imshow(out_img)
```

```python
omg = imread('https://stepik.org/media/attachments/lesson/61041/median-tiger.png')
print(omg)
imshow(omg)
np.array_equal(omg,out_img)
```
